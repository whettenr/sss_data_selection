```python
durations = {}

for name, path in csv_files.items():
    try:
        # Only read the duration column
        df = pd.read_csv(path, usecols=["duration"])
        if name == "large_len_split": 
            new_durations = pd.concat(
                [df["duration"].dropna() / 2, df["duration"].dropna() / 2],
                ignore_index=True
            )
            durations[name] = new_durations
        else:
            durations[name] = df["duration"].dropna()
        
        print(f"{name}:\t{(durations[name].sum() / 3600).round(0)}")
            
    except ValueError:
        print(f"⚠️ No 'duration' column found in {name}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Use Times New Roman (or Times fallback if not available)
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman", "Times", "DejaVu Serif"]

# --- Boxplot ---
plt.figure(figsize=(12, 5))

ticks = ["Small\n(250 hours)", "Dev\n(16 hours)", "Test\n(17 hours)",
         "Large\n(25k hours)", "Random\n(12.6k hours)",
         "MFCC\n(12.6k hours)", "Speaker\n(12.6k hours)",
         "SENSE\n(12.6k hours)", "Length\n(12.6k hours)", "Speaker+Len\n(12.5k hours)",
        "Length Split\n(12.5k hours)"]

box = plt.boxplot(
    durations.values(),
    tick_labels=ticks,
    showfliers=False,
    patch_artist=True,    # allows colored + hatched boxes
    medianprops=dict(color="black", linewidth=1.5),
    whiskerprops=dict(color="gray", linewidth=1.2),
    capprops=dict(color="gray", linewidth=1.2),
)

# --- Add Mean Markers ---
means = [np.mean(data) for data in durations.values()]
x_positions = range(1, len(ticks) + 1)
plt.plot(x_positions, means, 'D', markerfacecolor='black', markeredgecolor='black', markersize=5, label='Mean')

# Colors from Okabe–Ito colorblind-friendly palette
highlight_color = "#E69F00"  # orange
default_color   = "#4C72B0"  # blue

# Apply colors + hatching
for i, patch in enumerate(box["boxes"]):
    if i < 3:  # first three
        patch.set_facecolor(highlight_color)
        patch.set_alpha(0.9)
        patch.set_hatch("//")  # diagonal hatching
    else:
        patch.set_facecolor(default_color)
        patch.set_alpha(0.7)

# Legend (color + hatch aware)
highlight_patch = mpatches.Patch(facecolor=highlight_color, hatch="//", alpha=0.9,
                                 label="Fine-Tuning Data")
default_patch   = mpatches.Patch(facecolor=default_color, alpha=0.7,
                                 label="Pre-training Data")
plt.legend(handles=[highlight_patch, default_patch, plt.Line2D([0], [0], marker='D', color='w', markerfacecolor='black', markersize=5, label='Mean')],
           loc="upper right", frameon=True)

plt.xticks(rotation=30, ha="center", fontsize=10)
plt.ylabel("Duration (seconds)", fontsize=12)
plt.title("Distribution of Utterance Lengths", fontsize=14, weight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
# plt.savefig("length_distribution_all.pdf", format="pdf", dpi=300)
plt.show()
```

```python
baseline = durations["small_train"]

# Histogram settings
bins = range(0, 61, 1)  # 0–60 seconds, 1s bins

# Create one plot per comparison
for name, values in durations.items():
    if name == "small_train":
        continue  # skip comparing with itself

    plt.figure(figsize=(8, 5))
    
    # Baseline histogram (always in blue)
    plt.hist(baseline, bins=bins, alpha=0.5, label="small_train", density=True, color="blue")
    
    # Overlay current dataset (in red)
    plt.hist(values, bins=bins, alpha=0.5, label=name, density=True, color="red")
    
    plt.xlim(0, 60)
    plt.xlabel("Duration (seconds)")
    plt.ylabel("Density")
    plt.title(f"Duration Histogram: small_train vs {name}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
